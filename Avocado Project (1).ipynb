```python
#scaling the data with min  max scaler 
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
```

```python
lr=LinearRegression()
for i in range(0,100):
    x_train,x_test,y_train,y_test,=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    print(f"At random state {i},the training accuracy is = {r2_score(y_train,pred_train)}")
    print(f"At random state {i},the testing accuracy is = {r2_score(y_test,pred_test)}")
    print("\n")
```

Moving foward, we'll be doing a cross validation to ensure there no overfitting before moving on to hyperparameter tuning and auc-roc curve.

```python
#Cross validation  for decision tree
scr=cross_val_score(lr,x,y,cv=5)
scr.mean()
print("Cross validation score of DecisionTree:",scr.mean())
```

It is overfitted which is fine

```python
#regularising the data 
parameters={'criterion':['mse','mae'],'max_features':['auto','sqrt','log2']}
ls=DecisionTreeRegressor()
clf=GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)
```

```python
print(clf.best_params_)
```

```python
ds=DecisionTreeRegressor(criterion ="mse", max_features= "auto")
ds.fit(x_train,y_train)
ds.score(x_train,y_train)
pred_decision=ds.predict(x_test)
```

```python
rfs=r2_score(y_test,pred_decision)
print('Cross validation score:',rfs*100)
```

Now,let' save the model. I have chosen the DecisonTreeRegressor due to the fact that it gives the best model that works with real life problems. Now I'll move to region.

```python
#saving the model
import pickle 
filename='churn.pkl'
pickle.dump(ds,open(filename,'wb'))
```
